# Reading fits data

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as clr
import dfitspy as dft
from astropy.io import fits
import os
```

```python
p1 = os.getcwd()

h1 = dft.get_files(['all'],  dire=p1 + '/Spectro-Module/raw/')
listkeys = ['HIERARCH ESO INS MODE', 'HIERARCH ESO DPR TYPE', 'HIERARCH ESO OBS TARG NAME', 'EXPTIME']
fitsort = dft.dfitsort(h1, listkeys)

for key, value in fitsort.items():
    print(key + '\t')
    for j1,k1 in value.items():
        print(j1 + '\t' + k1 + '\n')
```

```python
hdul = fits.open(p1 + '/Spectro-Module/raw/HARPS.2006-09-07T22:04:47.576.fits')
h1 = hdul[1].data
h2 = np.transpose(h1)
plt.figure(figsize=(16*1.5,9*1.5))
plt.imshow(h2, origin='lower', norm=clr.LogNorm(),cmap='PuBu_r')
```
